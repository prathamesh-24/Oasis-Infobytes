<h3 style='color:orange'>Importing necessary machine learning libraries/h3>

<h5 style='color:red'>Standardized X with the help of StandardScaler</h5>

<h5 style='color:red'>Creating a logistic regression model</h5>

```python
model=LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

<h5 style='color:red'>Prediction</h5>

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
print(X_test.shape)
print(y_pred.shape)
```

<h5 style='color:red'>Training accuracy</h5>

```python
train_accuracy = model.score(X_train,Y_train)
```

```python
print("The training accurac is", train_accuracy)
```

<h5 style='color:red'>Test accuracy</h5>

```python
test_accuracy = model.score(X_test, Y_test)
```

```python
print("The testing accuracy is", test_accuracy)
```
